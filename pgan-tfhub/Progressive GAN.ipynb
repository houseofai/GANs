```python
import matplotlib.pyplot as plt
import tensorflow as tf
import tensorflow_hub as hub
import numpy as np
```

```python
model = "https://tfhub.dev/google/progan-128/1"

# 512 -> 128px
# 2048 -> 256px
# 8192 -> 512px
latent_dim = 512*4
size=256
latent_vector = tf.random.normal([1, latent_dim], seed=1338)
```

```python
hub_layer = hub.KerasLayer(model, input_shape=[], trainable=False)
```

```python
img = hub_layer(latent_vector)
print(latent_dim)
print(np.sqrt(img.shape[0]*img.shape[1]*img.shape[2]))
img = img.numpy().reshape(size,size,3)
plt.imshow(img)
plt.show()
```
